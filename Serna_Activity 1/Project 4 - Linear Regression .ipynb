```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('Datasets\\dengue.csv')
```

<h3>Is there a linear relationship between Year and Dengue Cases?</h3>

```python
df_year = dengue.groupby("Year")["Dengue_Cases"].sum().reset_index()

X = df_year["Year"].values.reshape(-1,1)
y = df_year["Dengue_Cases"].values.reshape(-1,1)

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

r2 = r2_score(y, y_pred)
r2
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.title("Year vs Dengue Cases")
plt.xlabel("Year")
plt.ylabel("Dengue Cases")
plt.show()
```

<h4>INSIGHTS: There is no clear linear relationship between Year and Dengue Cases. The trend line is almost flat, and the RÂ² value is very close to zero, which means the year does not explain or predict the number of dengue cases.</h4>

<h3>Is there a linear relationship between Year and Dengue Deaths?</h3>

```python
df_year = dengue.groupby("Year")["Dengue_Deaths"].sum().reset_index()

X = df_year["Year"].values.reshape(-1,1)
y = df_year["Dengue_Deaths"].values.reshape(-1,1)

model.fit(X, y)
y_pred = model.predict(X)
r2_score(y, y_pred)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.title("Year vs Dengue Deaths")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.show()
```

<h4>INSIGHTS:There is a clear linear relationship between Year and Dengue Deaths. The trend line shows a strong downward pattern, meaning dengue deaths steadily decreased from 2016 to 2020. The points follow the line much more closely compared to the dengue cases graph, showing a stronger linear trend.</h4>

<h3>Does dengue cases increase over the years?</h3>

```python
x = dengue["Year"].values.reshape(-1,1)
y = dengue["Dengue_Cases"].values.reshape(-1,1)

model = LinearRegression()
model.fit(x,y)

y_pred = model.predict(x)
r2 = r2_score(y, y_pred)

plt.scatter(x,y)
plt.plot(x,y_pred)
plt.title("Year vs Dengue Cases")
plt.xlabel("Year")
plt.ylabel("Dengue Cases")
plt.show()

print("RÂ²:", r2)
```

<h4>INSIGHTS: Dengue cases do not steadily increase over the years but instead go up and down. Although 2019 has the highest number of dengue cases compared to the other years, the earlier years show much lower and varying values. This means there is no clear upward trend from 2016 to 2020, and the changes in dengue cases are likely influenced by other factors such as weather, environment, and prevention efforts rather than just the passage of time.</h4>
